# **Project Summary -**

Customer Satisfaction Intelligence Platform is an intelligent Machine Learning
based system developed to predict customer satisfaction levels before customer feedback is received. The proposed solution analyzes customer interaction details such as communication channels, product categories, customer locations, handling time, agent information and purchase details to estimate Customer Satisfaction (CSAT) scores.

Unlike conventional approaches that rely solely on post-interaction feedback collection, the system enables proactive customer experience management by identifying potentially dissatisfied customers at an early stage. The project employs Random Forest based multi-class classification along with feature engineering, ordinal encoding and class imbalance handling techniques to improve predictive performance. Furthermore, an interactive multi-page analytical dashboard has been developed to provide feature importance analysis, prediction reports and business intelligence visualizations, making the solution suitable for real-world customer support environments.

## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Hypothesis testing is performed to validate statistically significant relationships between important customer support attributes and Customer Satisfaction (CSAT) scores. The following hypotheses are formulated based on the insights obtained from exploratory data analysis and visualization experiments conducted on the dataset.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H₀): There is no statistically significant relationship between Connected Handling Time and Customer Satisfaction (CSAT) scores.

Alternative Hypothesis (H₁): There is a statistically significant relationship between Connected Handling Time and Customer Satisfaction (CSAT) scores.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value

from scipy.stats import pearsonr

# Performing Pearson Correlation Test
correlation_coefficient, p_value = pearsonr(
    df["connected_handling_time"],
    df["CSAT Score"]
)

print("Correlation Coefficient :", correlation_coefficient)
print("P-Value :", p_value)
````

##### Which statistical test have you done to obtain P-Value?

Pearson Correlation Test was performed to determine whether a statistically significant relationship exists between Connected Handling Time and Customer Satisfaction (CSAT) scores.

##### Why did you choose the specific statistical test?

The Pearson Correlation Test was selected because both Connected Handling Time and Customer Satisfaction (CSAT) scores are numerical variables. The test effectively measures the strength and direction of the linear relationship between these variables while providing a P-value to evaluate statistical significance.

````python
# Perform Statistical Test to obtain P-Value

from scipy.stats import f_oneway

# Creating groups based on Product Category
groups = [
    group["CSAT Score"].values
    for _, group in df.groupby("Product_category")
]

# Performing One-Way ANOVA Test
f_statistic, p_value = f_oneway(*groups)

print("F-Statistic :", f_statistic)
print("P-Value :", p_value)
````

##### Which statistical test have you done to obtain P-Value?

One-Way ANOVA (Analysis of Variance) Test was performed to determine whether Customer Satisfaction (CSAT) scores differ significantly across different Product Categories.

The One-Way ANOVA Test was selected because Product Category is a categorical variable consisting of multiple groups, whereas Customer Satisfaction (CSAT) Score is a numerical variable. ### 9. Handling Imbalanced Dataset

##### Do you think the dataset is imbalanced? Explain Why.

Yes, the dataset exhibited class imbalance in Customer Satisfaction (CSAT) score distributions, where certain satisfaction classes contained substantially more observations than others. Class imbalance may bias Machine Learning models toward majority classes, leading to poor predictive performance for minority classes. Addressing this issue is particularly important for customer satisfaction prediction because accurately identifying less frequent satisfaction levels is essential for proactive customer experience management and business decision-making.

````python
# Handling Imbalanced Dataset

from imblearn.over_sampling import SMOTE

# Applying SMOTE on training data
smote = SMOTE(
    sampling_strategy="auto",
    random_state=42
)

X_train_smote, y_train_smote = smote.fit_resample(
    X_train,
    y_train
)

print("Original Training Shape :", X_train.shape)
print("Balanced Training Shape :", X_train_smote.shape)
````

##### What technique did you use to handle the imbalance dataset and why? (If needed to be balanced)

Synthetic Minority Over-sampling Technique (SMOTE) was employed to address class imbalance within the training dataset. SMOTE generates synthetic samples for minority classes by creating new observations based on existing feature patterns rather than simply duplicating records. This technique improves the model's ability to learn meaningful representations from underrepresented classes and contributes to balanced predictive performance across all Customer Satisfaction (CSAT) categories. SMOTE was applied exclusively to the training dataset to prevent information leakage and ensure unbiased model evaluation on unseen testing data.

````python
# Hyperparameter Optimization

from sklearn.model_selection import RandomizedSearchCV

param_grid = {
    "n_estimators":[100,200,300],
    "max_depth":[10,20,None],
    "min_samples_split":[2,5],
    "min_samples_leaf":[1,2],
    "bootstrap":[True,False]
}

random_search = RandomizedSearchCV(
    estimator=RandomForestClassifier(
        random_state=42
    ),
    param_distributions=param_grid,
    cv=5,
    n_iter=10,
    scoring="accuracy",
    random_state=42,
    n_jobs=-1
)

# Fit the Algorithm
random_search.fit(
    X_train_smote,
    y_train_smote
)

# Predict on the model
best_model = random_search.best_estimator_

y_pred_tuned = best_model.predict(
    X_test
)
````
